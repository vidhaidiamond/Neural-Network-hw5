```python
import pandas as pd

# Create a tiny fake dataset
data = pd.DataFrame({
    'score': [1, 0, 1, 0, 1, 0, 1, 0],  # Model's predictions (1=positive, 0=negative)
    'label_value': [1, 0, 1, 1, 0, 0, 1, 1],  # Ground truth (1=positive, 0=negative)
    'race': ['white', 'white', 'black', 'black', 'black', 'white', 'black', 'white']  # Group attribute
})

data
```

```python
from aequitas.group import Group
from aequitas.bias import Bias

# Analyze group metrics
g = Group()
xtab, _ = g.get_crosstabs(data)

# Analyze bias
b = Bias()
# Pass the original DataFrame and reference group dictionary to get_disparity_predefined_groups
bias = b.get_disparity_predefined_groups(xtab, original_df=data, ref_groups_dict={'race': 'white'})

# Display bias table
# Accessing relevant columns from the bias DataFrame
bias[['attribute_name', 'attribute_value', 'fnr', 'fpr_disparity', 'fnr_disparity']]
```
